```python
import pandas as pd
data = {
'Customer_ID':[101,102,103,104,104,105],
'Age':[23,25,None,30,30,28],
'City':['Delhi','Mumbai','Delhi',None,None,'Hyderabad'],
'Sales':[1500,2000,1800,2200,2200,None]
}
df = pd.DataFrame(data)
df.to_csv(
'raw_data.csv',
index=False
)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df['Age'] = df['Age'].fillna(
df['Age'].mean()
)
df['Sales'] = df['Sales'].fillna(
df['Sales'].mean()
)
df['City'] = df['City'].fillna(
"Unknown"
)
df
```

```python
report = df.describe()
report
```

```python
import matplotlib.pyplot as plt
df['Sales'].plot()
plt.title(
"Sales Summary"
)
plt.show()
```

```python
df.to_csv(
'cleaned_data.csv',
index=False
)

print('Cleaned file created successfully')
```
